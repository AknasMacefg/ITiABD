2. Решить задачу 1, распараллелив вычисления с помощью модуля `multiprocessing`. Для обработки каждого файла создать свой собственный процесс. 

````python
%%file out_string_.py

def counter(filename, output):
    with open(filename, 'r') as file:
        text = file.read().replace('\n', '').replace('\xa0', ' ')
    dictionary = {}
    for i in text:
        if i.lower() not in dictionary:
            dictionary[i.lower()] = 1
        else:
            dictionary[i.lower()] = dictionary[i.lower()] + 1 
    output.put(dictionary)
````

````python
import out_string_
````

````python
import multiprocessing as mp

output = mp.Queue()
processes = [mp.Process(target=out_string_.counter, args=('10_multiprocessing_data/Dostoevskiy Fedor. Igrok - BooksCafe.Net.txt', output)), 
             mp.Process(target=out_string_.counter, args=('10_multiprocessing_data/Dostoevskiy Fedor. Prestuplenie i nakazanie - BooksCafe.Net.txt', output))]


for p in processes:
    p.start()

for p in processes:
    p.join()
    
results = [output.get() for p in processes]
print(results)
````

2. Напишите функцию, которая принимает на вход название файла, созданного в результате решения задачи 1, считает среднее значение количества шагов для каждого тэга и возвращает результат в виде словаря.

````python
def avg_count(filename):
    tags_dictionary = {}
    data = pd.read_csv(filename, decimal=';')
    for i, k in zip(data['tags'], data['n_steps']):
        tags_clean = i.strip('[]').split(', ')
        for j in tags_clean:
            if j not in tags_dictionary:
                tags_dictionary[j] = [1, k, 0]
            else:
                tags_dictionary[j][0] = tags_dictionary[j][0] + 1
                tags_dictionary[j][1] = tags_dictionary[j][1] + k
                tags_dictionary[j][2] = tags_dictionary[j][1] / tags_dictionary[j][0]

    return tags_dictionary

````

4. Решите задачу 3, распараллелив вычисления с помощью модуля `multiprocessing`. Для обработки каждого файла создайте свой собственный процесс. Определите, за какое время задача решается для всех файлов.

````python
%%file avg_count_.py
import pandas as pd
def avg_count(filename, output):
    tags_dictionary = {}
    data = pd.read_csv(filename, decimal=';')
    for i, k in zip(data['tags'], data['n_steps']):
        tags_clean = i.strip('[]').split(', ')
        for j in tags_clean:
            if j not in tags_dictionary:
                tags_dictionary[j] = [1, k, 0]
            else:
                tags_dictionary[j][0] = tags_dictionary[j][0] + 1
                tags_dictionary[j][1] = tags_dictionary[j][1] + k
                tags_dictionary[j][2] = tags_dictionary[j][1] / tags_dictionary[j][0]

    output.put(tags_dictionary)
````

````python
import avg_count_
````

````python
import multiprocessing as mp

def tags_count_merge():
    tags_avg = {}
    output = mp.Queue()
    processes = [mp.Process(target=avg_count_.avg_count, args=('id_tag_nsteps_' + str(k) + '.csv', output)) for k in range(0, 8)]
    
    for p in processes:
        p.start()
    
    results = [output.get() for p in processes]
    
    for p in processes:
        p.join()
    
    
    for new_tags in results:
        if tags_avg == {}:
            tags_avg = new_tags
        else:
            for k in new_tags.keys():
                if k not in tags_avg:
                    tags_avg[k] = new_tags[k]
                else:
                    tags_avg[k][0] = tags_avg[k][0] + new_tags[k][0]
                    tags_avg[k][1] = tags_avg[k][1] + new_tags[k][1]
                    tags_avg[k][2] = tags_avg[k][1] / tags_avg[k][0]
    
    print(tags_avg)
````

5. (*) Решите задачу 3, распараллелив вычисления с помощью модуля `multiprocessing`. Создайте фиксированное количество процессов (равное половине количества ядер на компьютере). При помощи очереди передайте названия файлов для обработки процессам и при помощи другой очереди заберите от них ответы. 

````python
%%file worker_.py
import avg_count_
def worker(input_queue, output_queue):
    """Функция-обработчик для каждого процесса"""
    while True:
        filename = input_queue.get()
        
        if filename is None:
            break
        
        avg_count_.avg_count(filename, output_queue)
````

````python
import multiprocessing as mp
import os


def tags_count_merge():
    num_cores = mp.cpu_count()
    num_processes = max(1, num   _cores // 2)
    
    input_queue = mp.Queue()
    output_queue = mp.Queue()
    
    file_list = ['id_tag_nsteps_' + str(k) + '.csv' for k in range(0, 8)]
    
    for filename in file_list:
        input_queue.put(filename)
    
    for _ in range(num_processes):
        input_queue.put(None)
    
    processes = []
    for _ in range(num_processes):
        p = mp.Process(target=worker_.worker, args=(input_queue, output_queue))
        p.start()
        processes.append(p)
    
    results = []
    for _ in range(len(file_list)):
        result = output_queue.get()
        results.append(result)
    
    for p in processes:
        p.join()
    
    tags_avg = {}
    for new_tags in results:
        if not tags_avg: 
            tags_avg = new_tags
        else:
            for tag, stats in new_tags.items():
                if tag not in tags_avg:
                    tags_avg[tag] = stats
                else:
                    tags_avg[tag][0] += stats[0]
                    tags_avg[tag][1] += stats[1]
                    tags_avg[tag][2] = tags_avg[tag][1] / tags_avg[tag][0]
    
    print(tags_avg)


if __name__ == "__main__":
    tags_count_merge()
````
