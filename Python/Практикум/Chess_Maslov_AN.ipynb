```python
import traceback
from IPython.display import clear_output
import sys

class ChessField():
    def __init__(self, PlayerWhite, PlayerBlack):
        self.gameMap = [[' ', '|', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', '|', ' '],
                        ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
                       ['8', '|', ' ', '█', ' ', '█', ' ', '█', ' ', '█', '|', '8'],
                       ['7', '|', '█', ' ', '█', ' ', '█', ' ', '█', ' ', '|', '7'],
                       ['6', '|', ' ', '█', ' ', '█', ' ', '█', ' ', '█', '|', '6'],
                       ['5', '|', '█', ' ', '█', ' ', '█', ' ', '█', ' ', '|', '5'],
                       ['4', '|', ' ', '█', ' ', '█', ' ', '█', ' ', '█', '|', '4'],
                       ['3', '|', '█', ' ', '█', ' ', '█', ' ', '█', ' ', '|', '3'],
                       ['2', '|', ' ', '█', ' ', '█', ' ', '█', ' ', '█', '|', '2'],
                       ['1', '|', '█', ' ', '█', ' ', '█', ' ', '█', ' ', '|', '1'],
                        ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
                       [' ', '|', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', '|', ' ']]
        self.gameMap[3] = ['7', '|'] + PlayerBlack.figures[0] + ['|', '7']
        self.gameMap[2] = ['8', '|'] + PlayerBlack.figures[1] + ['|', '8']
        self.gameMap[8] = ['2', '|'] + PlayerWhite.figures[0] + ['|', '2']
        self.gameMap[9] = ['1', '|'] + PlayerWhite.figures[1] + ['|', '1']
        self.__gameTrack = 1
        
    def UpdateMap(self, playerSide, playerChoice, startChoice):
        printMap = ''
        self.dangerArr = []
        for i in self.gameMap:
            for j in i: 
                if type(j) in Figure.__subclasses__():
                    if j.figureSide != playerSide and playerChoice == False:
                        self.dangerArr = self.EnemyYourFiguresMoveChecker([i.index(j), self.gameMap.index(i)], j, self.dangerArr)
                    elif [i.index(j), self.gameMap.index(i)] == [startChoice[0], startChoice[1]] and playerChoice == True:
                        self.dangerArr = self.EnemyYourFiguresMoveChecker([startChoice[0], startChoice[1]], j, self.dangerArr)
        i2 = 0
        j2 = 0
        for i in self.gameMap: 
            for j in i:
                if type(j) in Figure.__subclasses__():
                    if j2 == startChoice[0] and i2 == startChoice[1]:
                        printMap = printMap + "\033[33m{}".format(j.figureMask) + '\033[0m '
                    elif [i2, j2] in self.dangerArr and j.figureSide != playerSide:
                        printMap = printMap + "\033[31m{}".format(j.figureMask) + '\033[0m '                 
                    else:
                        printMap = printMap + j.figureMask + ' '
                       
                else:
                    if playerChoice == True:
                        if [i2, j2] not in self.dangerArr:
                            printMap = printMap + str(j) + ' '
                        else:
                            printMap = printMap + "\033[33m\033[43m{}".format(str(j)) + '\033[0m '
                    else:
                        printMap = printMap + str(j) + ' '
                j2 += 1
            printMap = printMap + '\n'
            j2 = 0
            i2 += 1
        print(printMap)
        print("Ход: ", self.__gameTrack)
        
    
    def MovingFigure(self, choiceStart, choiceEnd, side, otherPlayer):
        if self.__MovementLockCheck(choiceStart, choiceEnd, side) == True:
            if type(self.gameMap[choiceEnd[1]][choiceEnd[0]]) in  Figure.__subclasses__():
                if self.gameMap[choiceEnd[1]][choiceEnd[0]].figureSide == side:
                    clear_output()
                    print("Нельзя атаковать свою фигуру!")
                    return False
                if isinstance(self.gameMap[choiceEnd[1]][choiceEnd[0]], Pawn):
                    otherPlayer.figures[0].remove(self.gameMap[choiceEnd[1]][choiceEnd[0]])
                else:
                    otherPlayer.figures[1].remove(self.gameMap[choiceEnd[1]][choiceEnd[0]])
            self.gameMap[choiceEnd[1]][choiceEnd[0]] = self.gameMap[choiceStart[1]][choiceStart[0]]
            self.__gameTrack += 1
            if (choiceStart[1] % 2 != 0  and choiceStart[0] % 2 == 0) or (choiceStart[1] % 2 == 0 and choiceStart[0] % 2 != 0):
                self.gameMap[choiceStart[1]][choiceStart[0]] = "█"
            else:
                self.gameMap[choiceStart[1]][choiceStart[0]] = " "
            clear_output()
            return True
        return False

    def EnemyYourFiguresMoveChecker(self, choiceStart, figure, dangerArr):
        for i in figure.moveRules.keys():
            if i == 'up' and isinstance(figure, Pawn) == False:
                for q in range(0, figure.moveRules[i]):
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] - q <= 2:
                        break
                    elif type(self.gameMap[choiceStart[1] - q][choiceStart[0]]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - q][choiceStart[0]].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - q, choiceStart[0]])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] - q, choiceStart[0]])
                    
            elif i == 'up-right':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] - q < 2 or choiceStart[0] + q > 9:
                        break
                    elif type(self.gameMap[choiceStart[1] - q][choiceStart[0] + q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - q][choiceStart[0] + q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - q, choiceStart[0] + q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] - q, choiceStart[0] + q])    
            elif i == 'right':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[0] + q > 9:
                        break
                    elif type(self.gameMap[choiceStart[1]][choiceStart[0] + q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1]][choiceStart[0] + q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1], choiceStart[0] + q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1], choiceStart[0] + q])
            elif i == 'down-right':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] + q > 9 or choiceStart[0] + q > 9:
                        break
                    elif type(self.gameMap[choiceStart[1] + q][choiceStart[0] + q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + q][choiceStart[0] + q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + q, choiceStart[0] + q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] + q, choiceStart[0] + q])
            elif i == 'down' and isinstance(figure, Pawn) == False:
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] + q > 9:
                        break
                    elif type(self.gameMap[choiceStart[1] + q][choiceStart[0]]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + q][choiceStart[0]].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + q, choiceStart[0]])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] + q, choiceStart[0]])
            elif i == 'down-left':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] + q > 9 or choiceStart[0] - q < 2:
                        break
                    elif type(self.gameMap[choiceStart[1] + q][choiceStart[0] - q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + q][choiceStart[0] - q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + q, choiceStart[0] - q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] + q, choiceStart[0] - q])
            elif i == 'left':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[0] - q < 2:
                        break
                    elif type(self.gameMap[choiceStart[1]][choiceStart[0] - q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1]][choiceStart[0] - q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1], choiceStart[0] - q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1], choiceStart[0] - q])
            elif i == 'up-left':
                for q in range(0, figure.moveRules[i]): 
                    if q == 0:
                        q = q + 1
                    if choiceStart[1] - q < 2 or choiceStart[0] - q < 2:
                        break
                    elif type(self.gameMap[choiceStart[1] - q][choiceStart[0] - q]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - q][choiceStart[0] - q].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - q, choiceStart[0] - q])
                            break
                        else:
                            break
                    else:
                        dangerArr.append([choiceStart[1] - q, choiceStart[0] - q])
            elif i == 'exotic':
                if choiceStart[1] - 2 >= 2 and choiceStart[0] + 1 <= 9:
                    if type(self.gameMap[choiceStart[1] - 2][choiceStart[0] + 1]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - 2][choiceStart[0] + 1].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - 2, choiceStart[0] + 1])
                    else:
                        dangerArr.append([choiceStart[1] - 2, choiceStart[0] + 1])
                if choiceStart[1] - 2 >= 2 and choiceStart[0] - 1 >= 2:
                    if type(self.gameMap[choiceStart[1] - 2][choiceStart[0] - 1]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - 2][choiceStart[0] - 1].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - 2, choiceStart[0] - 1])
                    else:
                        dangerArr.append([choiceStart[1] - 2, choiceStart[0] - 1])
                if choiceStart[1] - 1 >= 2 and choiceStart[0] + 2 <= 9:
                    if type(self.gameMap[choiceStart[1] - 1][choiceStart[0] + 2]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - 1][choiceStart[0] + 2].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - 1, choiceStart[0] + 2])
                    else:
                        dangerArr.append([choiceStart[1] - 1, choiceStart[0] + 2])
                if choiceStart[1] + 1 <= 9 and choiceStart[0] + 2 <= 9:
                    if type(self.gameMap[choiceStart[1] + 1][choiceStart[0] + 2]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + 1][choiceStart[0] + 2].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + 1, choiceStart[0] + 2])
                    else:
                        dangerArr.append([choiceStart[1] + 1, choiceStart[0] + 2])
                if choiceStart[1] + 2 <= 9 and choiceStart[0] - 1 >= 2:
                    if type(self.gameMap[choiceStart[1] + 2][choiceStart[0] - 1]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + 2][choiceStart[0] - 1].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + 2, choiceStart[0] - 1])
                    else:
                        dangerArr.append([choiceStart[1] + 2, choiceStart[0] - 1])
                if choiceStart[1] + 2 <= 9 and choiceStart[0] + 1 <= 9:
                    if type(self.gameMap[choiceStart[1] + 2][choiceStart[0] + 1]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + 2][choiceStart[0] + 1].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + 2, choiceStart[0] + 1])
                    else:
                        dangerArr.append([choiceStart[1] + 2, choiceStart[0] + 1])
                if choiceStart[1] + 1 <= 9 and choiceStart[0] - 2 >= 2:
                    if type(self.gameMap[choiceStart[1] + 1][choiceStart[0] - 2]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] + 1][choiceStart[0] - 2].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] + 1, choiceStart[0] - 2])
                    else:
                        dangerArr.append([choiceStart[1] + 1, choiceStart[0] - 2])
                if choiceStart[1] - 1 >= 2 and choiceStart[0] - 2 >= 2:
                    if type(self.gameMap[choiceStart[1] - 1][choiceStart[0] - 2]) in Figure.__subclasses__():
                        if self.gameMap[choiceStart[1] - 1][choiceStart[0] - 2].figureSide != figure.figureSide:
                            dangerArr.append([choiceStart[1] - 1, choiceStart[0] - 2])
                    else:
                        dangerArr.append([choiceStart[1] - 1, choiceStart[0] - 2])
        return dangerArr
            
    
    def __MovementLockCheck(self, choiceStart, choiceEnd, side):
        try:
            currentStartPoint = self.gameMap[choiceStart[1]][choiceStart[0]]
            currentEndPoint = self.gameMap[choiceEnd[1]][choiceEnd[0]]
            if type(currentStartPoint) not in Figure.__subclasses__():
                clear_output()
                print("Выбранная ячейка не имеет фигуры!")
                return False
            elif currentStartPoint.figureSide != side:
                clear_output()
                print("Выбранная фигура не вашего цвета!")
                return False
            elif (choiceEnd[1] < 2 or choiceEnd[1] > 10) or (choiceEnd[0] < 2 or choiceEnd[0] > 10):
                clear_output()
                print("Выбранная ячейка для перемещения выходит за рамки игрового поля!")
                return False
            elif (choiceEnd[1] == choiceStart[1] and choiceEnd[0] == choiceStart[0]):
                clear_output()
                print("Нельзя выбрать одну и ту же точку для перемещения!")
                return False
            else:
                rule = ''
                if choiceEnd[0] == choiceStart[0] and choiceEnd[1] < choiceStart[1]:
                    rule = 'up'
                elif choiceEnd[0] > choiceStart[0] and choiceEnd[1] < choiceStart[1] and (abs(choiceEnd[1] - choiceStart[1]) == abs(choiceEnd[0] - choiceStart[0])):
                    rule = 'up-right'
                elif choiceEnd[0] > choiceStart[0] and choiceEnd[1] == choiceStart[1]:
                    rule = 'right'
                elif choiceEnd[0] > choiceStart[0] and choiceEnd[1] > choiceStart[1] and (abs(choiceEnd[1] - choiceStart[1]) == abs(choiceEnd[0] - choiceStart[0])):
                    rule = 'down-right'
                elif choiceEnd[0] == choiceStart[0] and choiceEnd[1] > choiceStart[1]:
                    rule = 'down'
                elif choiceEnd[0] < choiceStart[0] and choiceEnd[1] > choiceStart[1] and (abs(choiceEnd[1] - choiceStart[1]) == abs(choiceEnd[0] - choiceStart[0])):
                    rule = 'down-left'
                elif choiceEnd[0] < choiceStart[0] and choiceEnd[1] == choiceStart[1]:
                    rule = 'left'
                elif choiceEnd[0] < choiceStart[0] and choiceEnd[1] < choiceStart[1] and (abs(choiceEnd[1] - choiceStart[1]) == abs(choiceEnd[0] - choiceStart[0])):
                    rule = 'up-left'
                    
                if currentStartPoint.MovingFigureRules(rule, self.gameMap, choiceStart, choiceEnd) == False:
                    return False
                else:
                    return True
            
        except:
            print(traceback.format_exc())
            clear_output()
            print("Введенное сообщение не соответствует требованию или выходит за рамки игрового поля!")
            return False
        
        
    

class Player():   
    def __init__(self, figures, side):
        self.figures = figures
        self.side = side
    
    def CurrentTurn(self):   
        try:
            print("Сторона: ", self.side)
            choiceStartIn = input("Выберите фигуру (напишите 0, чтобы сдаться): ")
            if choiceStartIn == '0':
                 return 0
            choiceStart = [Player.LetterChanger(choiceStartIn[0]), 12 - (int(choiceStartIn[1]) + 2)]
            clear_output()
            newGame.UpdateMap(self.side, True, choiceStart)
            choiceEndIn = input("Выберите куда поставить (напишите 0, чтобы отменить ход): ")
            if choiceEndIn == '0':
                clear_output()
                return 1
            choiceEnd = [Player.LetterChanger(choiceEndIn[0]), 12 - (int(choiceEndIn[1]) + 2)]
            if len(choiceStartIn) > 2 or len(choiceEndIn) > 2:
                raise ValueError
        except:
            return [-1, -1]
        return [choiceStart, choiceEnd]
        
    def LetterChanger(letter):
        match letter:
            case "A": return 2
            case "B": return 3
            case "C": return 4
            case "D": return 5
            case "E": return 6
            case "F": return 7
            case "G": return 8
            case "H": return 9
            case _: return -1
                
    
class Figure():
    def __init__(self, figureSide):
        self.figureSide = figureSide
    def MovingBetweenChecker(self, rule, gameMap, choiceStart, choiceEnd):
        tempcheck = False
        if rule == 'up':
            for i in range(1 , abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] - i][choiceStart[0]]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'up-right':
            for i in range(1, abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] - i] [choiceStart[0] + i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'right':
            for i in range(1 , abs(choiceEnd[0] - choiceStart[0])):
                if type(gameMap[choiceStart[1]][choiceStart[0] + i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'down-right':
            for i in range(1 , abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] + i] [choiceStart[0] + i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'down':
            for i in range(1 , abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] + i][choiceStart[0]]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'down-left':
            for i in range(1 , abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] + i] [choiceStart[0] - i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'left':
            for i in range(1 , abs(choiceEnd[0] - choiceStart[0])):
                if type(gameMap[choiceStart[1]][choiceStart[0] - i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        elif rule == 'up-left':
            for i in range(1 , abs(choiceEnd[1] - choiceStart[1])):
                if type(gameMap[choiceStart[1] - i] [choiceStart[0] - i]) in Figure.__subclasses__():
                    tempcheck = True
                    break
                                    
        if tempcheck == True:
            clear_output()
            print("Фигура не может проходить через других!")
            return False
        else:
            return True
            
        
class Pawn(Figure):
    def __init__(self, side, figureMask):
        if side == 'white':
            self.moveRules = {'up': 1, 'up-right': 1, 'up-left': 1}
        else:
            self.moveRules = {'down': 1, 'down-right': 1, 'down-left': 1}
        self.newMoving = True
        self.figureMask = figureMask
        super().__init__(side)
    
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if gameMap[choiceStart[1]][choiceStart[0]].newMoving == True and rule in self.moveRules.keys() and (abs(choiceEnd[1] - choiceStart[1]) == 2 or abs(choiceEnd[1] - choiceStart[1]) == 1):
            gameMap[choiceStart[1]][choiceStart[0]].newMoving = False
            return True
        elif self.moveRules[rule] == abs(choiceEnd[1] - choiceStart[1]) and type(gameMap[choiceEnd[1]][choiceEnd[0]]) not in Figure.__subclasses__():
            return True
        elif self.moveRules[rule] == abs(choiceEnd[0] - choiceStart[0]) and self.moveRules[rule] == abs(choiceEnd[1] - choiceStart[1]) and type(gameMap[choiceEnd[1]][choiceEnd[0]]) in Figure.__subclasses__():
            return True
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False
        
class Knight(Figure):
    def __init__(self, side, figureMask):
        self.moveRules = {'exotic': 4}
        self.figureMask = figureMask
        super().__init__(side)
        
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if (abs(choiceEnd[1] - choiceStart[1]) == 2 and abs(choiceEnd[0] - choiceStart[0]) == 1) or (abs(choiceEnd[1] - choiceStart[1]) == 1 and abs(choiceEnd[0] - choiceStart[0]) == 2):
            return True
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False
            
class Rook(Figure):
    def __init__(self, side, figureMask):
        self.moveRules = {'up': 8, 'left': 8, 'down': 8, 'right': 8}
        self.figureMask = figureMask
        super().__init__(side)
        
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if self.moveRules[rule] >= abs(choiceEnd[0] - choiceStart[0]) and self.moveRules[rule] >= abs(choiceEnd[1] - choiceStart[1]):
            if self.MovingBetweenChecker(rule, gameMap, choiceStart, choiceEnd):
                return True
            else:
                return False
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False
        
        
class King(Figure):
    def __init__(self, side, figureMask):
        self.moveRules = {'up': 1, 'up-right': 1, 'right': 1, 'down-right': 1, 'down': 1, 'down-left': 1, 'left': 1, 'up-left': 1}
        self.figureMask = figureMask
        super().__init__(side)
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if self.moveRules[rule] >= abs(choiceEnd[0] - choiceStart[0]) and self.moveRules[rule] >= abs(choiceEnd[1] - choiceStart[1]):
            if self.MovingBetweenChecker(rule, gameMap, choiceStart, choiceEnd):
                return True
            else:
                return False
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False

class Queen(Figure):
    def __init__(self, side, figureMask):
        self.moveRules = {'up': 8, 'up-right': 8, 'right': 8, 'down-right': 8, 'down': 8, 'down-left': 8, 'left': 8, 'up-left': 8}
        self.figureMask = figureMask
        super().__init__(side)
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if self.moveRules[rule] >= abs(choiceEnd[0] - choiceStart[0]) and self.moveRules[rule] >= abs(choiceEnd[1] - choiceStart[1]):
            if self.MovingBetweenChecker(rule, gameMap, choiceStart, choiceEnd):
                return True
            else:
                return False
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False

class Bishop(Figure):
    def __init__(self, side, figureMask):
        self.moveRules = {'up-right': 8, 'down-right': 8, 'down-left': 8, 'up-left': 8}
        self.figureMask = figureMask
        super().__init__(side)
    def MovingFigureRules(self, rule, gameMap, choiceStart, choiceEnd):
        if self.moveRules[rule] >= abs(choiceEnd[0] - choiceStart[0]) and self.moveRules[rule] >= abs(choiceEnd[1] - choiceStart[1]):
            if self.MovingBetweenChecker(rule, gameMap, choiceStart, choiceEnd):
                return True
            else:
                return False
        else:
            clear_output()
            print("Выбранная ячейка выходит за рамки возможностей фигуры!")
            return False



Player1 = Player([[Pawn("white", "P") for i in range(8)],[Rook("white", "R"), Knight("white", "N"), Bishop("white", "B"), Queen("white", "Q"), King("white", "K"), Bishop("white", "B"), Knight("white", "N"), Rook("white", "R"),]], "white")
Player2 = Player([[Pawn("black", "p") for i in range(8)],[Rook("black", "r"), Knight("black", "n"), Bishop("black", "b"), Queen("black", "q"), King("black", "k"), Bishop("black", "b"), Knight("black", "n"), Rook("black", "r"),]], "black")
newGame = ChessField(Player1, Player2)


while (True):
    try:
        while (True):
            newGame.UpdateMap(Player1.side, False, [0, 0])
            choice = Player1.CurrentTurn()
            if choice == 0:
                print(Player1.side, ' сдались!')
                raise SystemExit
            elif choice == 1:
                continue
            if newGame.MovingFigure(choice[0], choice[1], Player1.side, Player2) == True:
                break
        
        while (True):
            newGame.UpdateMap(Player2.side, False, [0, 0])
            choice = Player2.CurrentTurn()
            if choice == 0:
                print(Player2.side, ' сдались!')
                raise SystemExit0
            elif choice == 1:
                continue
            if newGame.MovingFigure(choice[0], choice[1], Player2.side, Player1) == True:
                break
    except:
        print(traceback.format_exc())
        print("Конец игры!")
        break
```
