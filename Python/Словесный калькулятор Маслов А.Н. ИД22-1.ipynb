```python
#Калькулятор слов
dic={"ноль":0,"один":1,"два":2, "одна":1, "две":2, "три":3,"четыре":4,"пять":5,"шесть":6,"семь":7,"восемь":8,"девять":9,"десять":10, 
     "одиннадцать":11,"двенадцать":12,"тринадцать":13,"четырнадцать":14,"пятнадцать":15,'шестнадцать':16,'семнадцать':17,
     'восемнадцать':18,"девятнадцать":19,"двадцать":20,"тридцать":30,"сорок":40,
     "пятьдесят":50,"шестьдесят":60,"семьдесят":70,"восемьдесят":80,"девяносто":90, "сто":100, "двести":200, "триста":300,
     "четыреста":400,"пятьсот":500,"шестьсот":600,"семьсот":700,"восемьсот":800,"девятьсот":900, "тысяча":1000,
    "две тысячи":2000, "три тысячи":3000, "четыре тысячи":5000, "пять тысяч":5000, "шесть тысяч":6000,
    "семь тысяч":7000, "восемь тысяч":8000, "девять тысяч":9000} #словарь значений

dicznak = {"плюс":'+', "минус":'-', "умножить":'*', "разделить":'/', "открывается":'(', "закрывается":')',
           "скобка":' ', 'на': ' '} #словарь знаков

dicfloat = {"и":'.', "десятых":0.1, "сотых":0.01, "тысячных":0.001, "десятая":0.1, "сотая":0.01, "тысячная":0.001} #словарь десятичных


print("Введите нужное выражение строчными буквами:")
while True: #проверка на правильный ввод
    leftbrakes = 0 #количество левых скобок
    rightbrakes = 0 #словарь правых скобок
    znaks = 0 #словарь знаков
    digits = 0 #словарь цифр
    zerodiv = False #проверка деления на ноль
    q = 0
    znak = True #Проверка на следующее значение
    op = [] #массив выражения
    floatnumber = 0 #количество цифр после запятой
    nextminus = False #проверка на цифру с минусом
    floatdigit = False #проверка на десятичное число
    lst=input().split() #разбиение вводимой строки на массив
   
    for i in lst:
        if dic.get(i) == None: #проверка на десятичное число
            if dicfloat.get(i) == '.':
                floatdigit = True
                continue
            if dicfloat.get(i) != None: #установка количества цифр после запятой
                floatnumber = floatnumber * dicfloat.get(i)
                if op[len(op) - 1] < 0:
                    op[len(op) - 1] = op[len(op) - 1] - floatnumber
                else:
                    op[len(op) - 1] = op[len(op) - 1] + floatnumber
                floatdigit = False
                continue
            
            if dicznak.get(i) == '-' and len(op) == 0: #проверка на следующий минус
                nextminus = True
                continue
            elif dicznak.get(i) == '-' and (op[len(op) - 1] == '-' or op[len(op) - 1] == '+' or op[len(op) - 1] == '*' or op[len(op) - 1] == '(' or op[len(op) - 1] == '/'):
                nextminus = True
                continue
                
            if dicznak.get(i) != ' ': #проверка на следующий знак
                q = q + 2
                op.append(dicznak.get(i))
                znak = True
                znaks = znaks + 1
            if dicznak.get(i) == '(':
                leftbrakes = leftbrakes + 1
                znaks = znaks - 1
            elif dicznak.get(i) == ')':
                znaks = znaks - 1
                rightbrakes = rightbrakes + 1
            continue
        
        if floatdigit == True: #запись десятичного числа
            floatnumber = floatnumber + dic.get(i)
            continue
        
        
        if znak == True: #запись обычного числа
            if nextminus == True:
                op.append(dic.get(i) * -1)
            else:
                op.append(dic.get(i))
            nextminus = False
            znak = False
            digits = digits + 1
            if op[len(op) - 2] == '/' and dic.get(i) == 0:
                zerodiv = True
        else:
            op[q] = op[q] + dic.get(i)
            digits = digits + 1
            if dic.get(i) == 0:
                zerodiv = True
    if None in op or q == 0 or znaks >= digits or leftbrakes != rightbrakes or zerodiv == True:
        print("Неправильно введено выражение! Попробуйте снова")
        continue
    else:
        break
    
            
def resh(end_ex, mass_ex): #рекурсиваня функция решения выражения
    while len(mass_ex) != 1:
        if '(' in mass_ex:
            checker = 0
            end_in = 0
            mass_in = []
            start_in = mass_ex.index('(')
            for i in range(start_in + 1, end_ex - 1):
                if mass_ex[i] == '(':
                    checker = checker + 1
                if mass_ex[i] == ')' and checker != 0:
                    checker = checker - 1
                elif mass_ex[i] == ')' and checker == 0:
                    end_in = i
                    break
                mass_in.append(mass_ex[i])
            mass_ex[start_in:end_in + 1] = resh(end_in, mass_in)
        else:
            while '*' in mass_ex or '/' in mass_ex:
                for i in mass_ex:
                    if i == '*':
                        mass_ex[mass_ex.index('*') - 1:mass_ex.index('*') + 2] = [mass_ex[mass_ex.index('*') - 1] * mass_ex[mass_ex.index('*') + 1]]
                        continue
                    elif i == '/':
                        mass_ex[mass_ex.index('/') - 1:mass_ex.index('/') + 2] = [mass_ex[mass_ex.index('/') - 1] / mass_ex[mass_ex.index('/') + 1]]
                        continue
            while '+' in mass_ex or '-' in mass_ex:
                for i in mass_ex:
                    if i == '+':
                        mass_ex[mass_ex.index('+') - 1:mass_ex.index('+') + 2] = [mass_ex[mass_ex.index('+') - 1] + mass_ex[mass_ex.index('+') + 1]]
                        continue
                    elif i == '-':
                        mass_ex[mass_ex.index('-') - 1:mass_ex.index('-') + 2] = [mass_ex[mass_ex.index('-') - 1] - mass_ex[mass_ex.index('-') + 1]]
                        continue
        
    return(mass_ex)

def get_name(d, value): #функция преобразования числа в слова
    res = []
    minus = False
    floatpart = 0
    j = 10
    
    if isinstance(value, float):  #проверка на тип float
        value = round(value, 3)
        floatpart = value % 1
        value = value - floatpart
        value = int(value)
    
    
    if (value < 0): #проверка на отрицательное число
        minus = True
        value = value * -1
        
    if (value > 19): #проверка на число больше 19
        while value != 0:
            if value % j == 0:
                j = j * 10
                continue
            res.append(value % j)
            value = value - value % j
            j = j * 10
    else:
        res.append(value)
          
    for i in res: #запись целого числа
        for k, v in d.items():
            if v == i:
                res[res.index(i)] = k
                break
    if minus == True:
        res.append("минус")
    res.reverse()
    
    if floatpart != 0: #запись десятичного числа
        res.append("и")
        round(floatpart, 3)
        floatstr = str(floatpart)
        floatstr = floatstr[2:5]
        
        if floatstr[1:3] == '00':
            floatstr = floatstr[0:1]
        if floatstr[2:3] == '0':
            floatstr = floatstr[0:2]
        
        floatpart = int(floatstr)
        j = 10
        massfloat = []
        if floatpart > 19:
            while floatpart != 0:
                if floatpart % j == 0:
                    j = j * 10
                    continue
                massfloat.append(floatpart % j)
                floatpart = floatpart - floatpart % j
                j = j * 10
        else:
            massfloat.append(floatpart)
        massfloat.reverse()
        
        for i in massfloat:
            for k, v in d.items():
                if v == i:
                    res.append(k)
                    break
        
        if len(floatstr) == 3:
            res.append("тысячных")
        elif len(floatstr) == 2:
            res.append("сотых")
        elif len(floatstr) == 1:
            res.append("десятых")
    return res
    
op = resh(len(op), op)
res = get_name(dic, op[0])
res_string = ''
for i in res:
    res_string = res_string + i + ' '
print("Ответ: " + res_string)


```
