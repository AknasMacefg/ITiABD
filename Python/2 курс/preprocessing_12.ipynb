### 8. Создайте конвейер для обработки количественных признаков, который будет осуществлять импутацию медианами и стандартизацию.

```python
numerical_pipe = Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
```

### 9. Создайте два конвейера для обработки категориальных признаков (номинальных и ординальных), которые осуществляют импутацию пропусков самой часто встречающейся категорией и преобразуют признаки в числовые при помощи классов OneHotEncoder и OrdinalEncoder соответственно.

```python
nominal_pipe = Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OneHotEncoder(sparse_output= False))])
ordinal_pipe = Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('encoder', OrdinalEncoder())])
```

### 10. Создайте список трансформеров –трехэлементных кортежей. Первый элемент кортежа – название конвейера с преобразованиями для определенного списка названий признаков, второй элемент – собственно конвейер, и третий элемент – соответствующий список названий признаков. Создайте экземпляр класса ColumnTransformer, передав в него список трансформеров. Трансформируйте (предобработайте) обучающие и тестовые данные и выведите их в виде датафреймов.

```python
transformers = [('numerical', numerical_pipe, numerical_columns), ('categorial_ordinal', ordinal_pipe, categorial_ordinals_columns), ('categorial_nominal', nominal_pipe, categorial_numinals_columns)]
transformer = ColumnTransformer(transformers=transformers)
X_train.transform = transformer.fit_transform(X_train)
X_train.transform
```

```python
transformer.fit(X_train)
```

```python
nominals_columns = transformer.named_transformers_['categorial_nominal'].named_steps['encoder'].get_feature_names_out(input_features = categorial_numinals_columns).tolist()
nominals_columns
```

```python
df_preprocess = pd.DataFrame(X_train.transform, columns = numerical_columns + categorial_numinals_columns + nominals_columns)
```

```python
df_preprocess.isna().sum()
```

```python
df_preprocess.head()
```

```python
X_test.transform = transformer.transform(X_test)
X_test.transform
```

```python
df_preprocess_test = pd.DataFrame(X_test.transform, columns = numerical_columns + categorial_numinals_columns + nominals_columns)
df_preprocess_test.isna().sum()
```

```python
df_preprocess_test.head()
```

1. Напишите класс Outlier, осуществляющий отбор выбросов по порогу. Добавьте в предобработку для признака Income обработку выбросов. Выбросами считайте доход равный нулю.

```python
class Outlier(): # класс уже создан
 """
 Удалние выбросов по порогу
 """
 def __init__(self, threshold= 0):
    self.threshold = threshold

 def fit(self, X, y=None):
    return self
 def transform(self, X):
    return np.where(X <= self.threshold, np.nan, X)
```

```python
outlier_income_pipe = Outlier()

numerical_income_pipe = Pipeline([
    ('outlier_income', outlier_income_pipe),
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
numerical_income_pipe
```

5. Повторите предобработку данных и обучение модели, используя make_pipeline, вместо Pipeline, и make_column_transformer, вместо ColumnTransformer.

```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer

# трансформер
preprocessor = make_column_transformer(
    (make_pipeline(Outlier(), SimpleImputer(strategy='median'), StandardScaler()), ['Income']),
    (make_pipeline(SimpleImputer(strategy='median'), StandardScaler()), numerical_columns_without_income),
    (make_pipeline(SimpleImputer(strategy='most_frequent'), OrdinalEncoder()), categorial_ordinals_columns),
    (make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(sparse_output=False, handle_unknown='ignore')), categorial_numinals_columns)
)

# используем make_pipeline
pipeline = make_pipeline(preprocessor, LogisticRegression(solver='lbfgs', max_iter=200))
pipeline.fit(X_train, y_train)
```

```python
# предобработка данных
X_train_transformed = pipeline[:-1].transform(X_train)
X_test_transformed = pipeline[:-1].transform(X_test)
feature_names = []

# 1. Numerical features (включая 'Income')
feature_names.extend(['Income'] + numerical_columns_without_income)

# 2. Ordinal features
feature_names.extend(categorial_ordinals_columns)

# 3. Nominal features (One-Hot Encoded)
onehot_encoder = preprocessor.named_transformers_['pipeline-4'].named_steps['onehotencoder']

# получаем названия признаков
nominal_features = list(onehot_encoder.get_feature_names_out(categorial_numinals_columns))
feature_names.extend(nominal_features)  # добавляем в список признаков


# преобразование в датафрейм
df_preprocessed_train = pd.DataFrame(X_train_transformed, columns=feature_names)
df_preprocessed_test = pd.DataFrame(X_test_transformed, columns=feature_names)
```
